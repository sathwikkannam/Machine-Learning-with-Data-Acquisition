```python
import pandas as pd
import import_ipynb as ipynb
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%%capture
m_knn = ipynb.NotebookLoader().load_module('movements_knn')
```

# Correlation
### One can use correlation to investigate the dependence between variables/features.

#### Calculate the correlation matrix between all features. From this result, discuss if all features are useful for the classification

```python
df : pd.DataFrame = m_knn.X
classes = m_knn.training_sets
```

```python
correlation_matrix = df.corr()
```

```python
plt.figure(figsize=(15, 10))
plt.title("All movement classes")
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.show()
```

### Calculate the correlation matrix between all features. From this result, discuss if all features are useful for the classification
#### Diagonal values are one as it is the correlation between the same two variables. 1 = positively correlated, -1 = negatively correlated, 0 = no correlation. Generally, these features are not correlated to each other as they are all near 0. So we need all features since they are not correlated at all.

```python
for i in classes:
    plt.title(i['class'][0])
    sns.heatmap(i.drop('class', axis='columns').corr(), annot=True, cmap='coolwarm', linewidths=.5)
    plt.show()
```
